## Create a new tidy DateFrame

```python
l_cols = df_agegen.columns.to_list()
unzip_cols = [cols.split(' - ') for cols in l_cols]
unzip_cols = unzip_cols[2:]
```

```python
unzip_cols
```

```python
len(unzip_cols)
```

```python
[ele[0] for ele in unzip_cols]
```

```python
df_new = pd.DataFrame(columns=['Cause','Gender', 'Age', 'Number'])
```

```python
len(df_agegen.loc[1].to_list()[2:])
```

```python
for i in range(len(df_agegen.index)):
    record = df_agegen.loc[i]
    cause = record['Cause']
    data = {
        'Cause': [cause for _ in range(24)],
        'Gender' : [ele[1] for ele in unzip_cols],
        'Age' : [ele[0] for ele in unzip_cols],
        'Number': df_agegen.loc[i].to_list()[2:]
    }
    df = pd.DataFrame(data)
    df_new = pd.concat([df_new, df], ignore_index=True)
```

```python
df_new
```

## Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df_new2 = df_new.copy()

df_new2['Cause'] = le.fit_transform(df_new2['Cause'])

df_new2['Gender'] = df_new2['Gender'].map({
    'Male': 1,
    'Female' : 2,
    'Transgender' : 3,
    'Total' : 0,
})

df_new2['Age'] = df_new2['Age'].map({
'Below 18 years': 1,
'18 and Above-Below 30 years' : 2,
'30 and Above-Below 45 years' : 3,
 '45 and Above-Below 60 years': 4,
 '60 years & Above' : 5,
 'Total' : 0,
})
```
